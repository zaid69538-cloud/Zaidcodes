```python
# first the class average
# find the highest score
# find the lowest high_score
#count the student above 75

import numpy as np
marks =np.array([78, 91, 55, 72, 40,
                 88,98,81,11,23,78,88,
                 91,96])
average = np.mean(marks)
highest_score = np.max(marks)
lowest_score = np.min(marks)
high_scores_count = np.sum(marks > 75)
print("Class Average:", average)
print("Highest Score:", highest_score)
print("Lowest Score:", lowest_score)
print("Number of Students above 75:", high_scores_count)
```

```python
import pandas as pd
import numpy as np

def get_result(avg):
  if avg>=40:
    return "Pass"
  return "Fail"

df = pd.read_csv("session4_dummy_students_data.csv.xls")

df["average"] = df[["math","science","english"]].mean(axis=1)
df["result"] = df["average"].apply(get_result)

print("Full_Data:")
print(df)

print("class average:",df["average"].mean())
print("Highest average:",df["average"].max())
print("Lowest average:",df["average"].min())

low_assignment_student = df[df["assignments_submitted"] < 6]
print("\nStudent with less than 6 assignments:")
print(low_assignment_student[["name","assignments_submitted"]])
```
